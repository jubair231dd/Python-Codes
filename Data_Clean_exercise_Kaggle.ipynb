## Scaling and Normalization

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.preprocessing import minmax_scaling
from scipy import stats

original_data = np.random.exponential(size = 1000)

scaled_data = minmax_scaling(original_data, columns = [0])

fig, ax = plt.subplots(1,2)
sns.histplot(original_data, ax = ax[0], kde= True , legend = False)
ax[0].set_title("Original Data")

sns.histplot(scaled_data, ax = ax[1], kde = True, legend = False)
ax[1].set_title('Scaled Data')
plt.show()
```

```python
normalized_data = stats.boxcox(original_data)

fig, ax = plt.subplots(1, 2)
sns.histplot(original_data, ax = ax[0], kde = True, legend = False)
ax[0].set_title("Original Data")

sns.histplot(normalized_data[0], ax = ax[1], kde = True, legend = False)
ax[1].set_title("Normalized Data")
plt.show()
```

```python
# df['name'] = df['Used name'].str.split('•').str[0] + ' ' + df['Used name'].str.split('•').str[1]
# df.head()
```

```python
df['name'] = df['Used name'].str.replace('•', ' ')
df.head()
```

```python
#splitting height and weight column

df['height_cm'] = df['Measurements'].str.split('/', expand = True)[0]
df['weight_kg'] = df['Measurements'].str.split('/', expand = True)[1]

df.head()
```

```python
# df['height'].str.contains('cm').sum()
df['height_cm'] = df['height_cm'].str.replace('cm', '')
df['weight_kg'] = df['weight_kg'].str.replace('kg', '')
df.head()
```

```python
df.info()
```

```python
df.drop(['Measurements', 'Used name'], axis = 1, inplace = True)
df.head()
```
