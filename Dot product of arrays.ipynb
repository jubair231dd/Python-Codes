```python
import pandas as pd
import numpy as np

# a = np.array([[5],[4],[3],[2],[1]])
# b = np.array([5,4,3,2,1,12,2,3,4,4])

def dotproduct(a,b):
    
    for items in range(len(b)):
        
    
        for i in np.take(b,[items]):
            c = np.array([[i]])
        #print(c.shape)
        print("a", "dot", "c", "\n")
        df = pd.DataFrame(a.dot(c))
        print(df, "\n")
        print("Minimum value is :", df[0].min())
        print("\n")
        
    
    
```

```python
a = np.array([[5],[4],[3],[2],[1]])
b = np.array([5,4,3,2,1,12,2,3,4,4])
```

```python
dotproduct(a,b)
```
